```python
import googletrends as googletrends
```

```python
# Gather temporal searches
results = googletrends.temporal(['Tellius', 'Tableau AskData'], geo=['US'], date_start='01-01-2020')

# Make plot using default settings
googletrends.plot(results, color_by_searchword=True, group_by_searchword=False)
```

```python
# Gather searches over geographical locations
results = googletrends.spatio(['Tellius','Microstrategy', 'Tableau'], geo=['US','NL','DE','italy','BE'], date_start='01-01-2020')

# Make plot
googletrends.plot(results)
```

```python
# Make worldmap plot
googletrends.plot_worldmap(results)
```

```python
# Gather searches over geographical locations
results = googletrends.trending(['corona','covid-19','virus'], geo=['NL','DE','italy','BE', 'US'], date_start='01-01-2020')

# Make plot
googletrends.plot(results)
```
